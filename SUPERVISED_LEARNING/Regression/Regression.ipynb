**Regression in machine learning is a supervised learning technique used to predict a continuous numerical value based on the relationship between independent and dependent variables. Unlike classification, which provides discrete class labels (e.g., "spam" or "not spam"), regression yields a numerical output (e.g., a house price or a person's age).**

---

## How Regression Works
Regression algorithms work by finding a "best-fit" line or curve that minimizes the difference (error) between the predicted output and the actual output in the training data.

1. `Labeled Data`: The model is trained on labeled data, meaning the input features (independent variables) and the correct continuous output values (dependent variable/target) are provided.

2. `Learning Relationships`: During training, the algorithm learns the specific mathematical relationship between the features and the target.

3. `Minimizing Error`: A cost function (commonly Mean Squared Error, or MSE) measures the prediction errors. Optimization algorithms, such as gradient descent, iteratively adjust the model's parameters (weights and bias) to minimize this cost.

4. `Prediction`: Once trained, the model can use this learned relationship to forecast or predict the continuous outcome for new, unseen input data.

---

## Common Regression Algorithms

Various algorithms are used for regression, depending on the complexity of the data and the desired outcome:

+ `Linear Regression`: The simplest form, which assumes a linear relationship between the independent and dependent variables and fits a straight line to the data.

+ `Multiple Linear Regression`: An extension of linear regression that uses multiple independent variables to predict the target variable.

+ `Polynomial Regression`: Used for modeling non-linear, curved relationships between variables by fitting an n-th degree polynomial function to the data.

+ `Ridge and Lasso Regression`: Regularization techniques that add a penalty to the model's coefficients to prevent overfitting, particularly useful when dealing with many features or multicollinearity.

+ `Decision Tree Regression`: A tree-like model that splits data into smaller subsets based on features, with the final prediction being the average value of the leaf node.

+ `Random Forest Regression`: An ensemble method that combines predictions from multiple decision trees to improve accuracy and stability.

+ `Support Vector Regression (SVR)`: An adaptation of Support Vector Machines that finds the optimal hyperplane while being robust to outliers and handling non-linear relationships effectively.

---

## Real-World Applications

Regression analysis is used across many industries to make data-driven decisions:

+ `Predicting Housing Prices`: Estimating a house's value based on features like size, location, and number of bedrooms.

+ `Forecasting Sales`: Predicting future sales volumes based on historical data, advertising spend, and seasonality.

+ `Medical Research`: Identifying risk factors for diseases and predicting patient recovery times or disease progression based on medical data.

+ `Financial Analysis`: Forecasting stock prices, assessing credit risks, and analyzing portfolio performance.

+ `Operational Efficiency`: Predicting system load or equipment failure times in technology and manufacturing.

```python
def plot_per_column(feature,target,column:int,model_pred:None):
    '''
    feature : Feature of the model (Independent Variable)
    target : Target of the model (Dependent Variable)
    column : Column of the feature
    model_pred : Prediction of the given model
    return : None | plot of feature and target variable with the given column
    '''
    plt.figure(figsize=(8,5))
    plt.scatter(feature[:,column],target,c="b",marker='o',label="Actual Data Point")
    #model_name = {"linear_model_pred":"Linear Model","ridge_model":"Rigde Model", "lasso_model":"Lasso Model", "svr_model":"SVM Regression Model","decision_tree_model":"Decision Tree Model","random_forest_model":"Random Forest Model","knn_model":"KNN Regression Model"}
    #for model_pred_key in model_name.keys():
    if model_pred is not None :
        plt.scatter(feature[:,column],model_pred,c='r',marker='^',label="Model Prediction Data Point")
    plt.title(f"Feature {column} vs Target")
    plt.xlabel(f"Feature {column}")
    plt.legend()
    plt.ylabel("Target")
    plt.show()
```

**Linear regression is a statistical and machine learning method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. This technique is used to predict continuous outcomes, such as house prices or test scores.
The fundamental goal of linear regression is to find a "best-fit" straight line (or a more complex surface in multiple dimensions) that minimizes the difference, or error (residuals), between the observed data points and the values predicted by the model. This is commonly achieved using the least squares method.**

---

## Key Concepts
+ `Dependent Variable (Y)`: The variable you want to predict or explain (also known as the response or target variable).

+ `Independent Variable(s) (X)`: The variable(s) used to make the prediction (also known as explanatory, predictor, or features).

+ `Linear Equation`: The relationship is represented by a linear equation, which for a simple model (one independent variable) is typically \(y=b+wx\) (or \(y=\beta _{0}+\beta _{1}x\)) where 'b' (or \(\beta _{0}\)) is the y-intercept and 'w' (or \(\beta _{1}\)) is the slope or coefficient.

+ `Coefficients`: These parameters are estimated during the model training process to define the best possible line or surface.

---

## Types of Linear Regression

1. `Simple Linear Regression`: Involves only one independent variable and one dependent variable.

2. `Multiple Linear Regression`: Involves one dependent variable and two or more independent variables.

---

## Applications
Linear regression is widely used across many fields:

+ `Business`: Forecasting sales or evaluating pricing elasticity.
+ `Finance`: Predicting stock prices based on economic indicators like interest rates.
+ `Healthcare`: Modeling the effects of diet and exercise on heart disease risk.
+ `Real Estate`: Estimating house prices based on features like size, location, and number of bedrooms.

**The LASSO model (Least Absolute Shrinkage and Selection Operator) is a popular regularization technique in machine learning and statistics that performs both variable selection and regularization to enhance the prediction accuracy and interpretability of a linear regression model. It is also known as L1 regularization.**

---

## Key Principles

1. `L1 Regularization`: Lasso regression adds a penalty term to the linear regression's loss function (residual sum of squares), proportional to the absolute value of the magnitude of the coefficients.
2. `Coefficient Shrinkage and Sparsity`: This penalty shrinks coefficients toward zero. A key characteristic of the Lasso model is its ability to force the coefficients of less important features to become exactly zero.

3. `Automatic Feature Selection`: By setting some coefficients to zero, Lasso automatically performs feature selection, effectively ignoring irrelevant variables and resulting in a simpler, more interpretable "sparse" model.

---

## How It Works
The objective function minimized by Lasso regression is:

    Minimize: Sum of Squared Residuals + λ(lambda) X (Sum of Absolute Values of Coefficients)

The λ (lambda) is the regularization parameter that controls the strength of the penalty.

+ If λ = 0, the model is equivalent to standard linear regression (OLS).

+ As λ increases, the penalty is stronger, forcing more coefficients to become exactly zero, which simplifies the model and increases bias while decreasing variance.

Choosing the optimal λ value, typically through cross-validation, is crucial to find the right balance between model simplicity and accuracy.

---

## When to Use a Lasso Model
The Lasso model is particularly useful in scenarios where:

1. You have a dataset with a large number of features, and you suspect that only a subset of those predictors is actually significant.

2. You need a model that is highly interpretable due to its ability to eliminate variables.

3. The dataset exhibits high multicollinearity, as it can help identify and select the most important features.

**The ridge model, or ridge regression, is a regularization technique used in statistics and machine learning to create more stable and generalizable linear regression models. It is particularly effective in addressing multicollinearity (high correlation between independent variables) and overfitting by introducing a penalty to the model's coefficients.**

---

## Key Principles
1. `L2 Regularization`: Ridge regression is known as L2 regularization because it adds a penalty term proportional to the square of the magnitude of the coefficients to the standard linear regression objective function (Ordinary Least Squares or OLS). This discourages the coefficients from becoming too large.

2. `Coefficient Shrinkage`: The penalty forces the model to shrink the regression coefficients towards zero. However, it never forces them to be exactly zero, meaning all features are retained in the model, albeit with reduced influence.

3. `Bias-Variance Trade-off`: By introducing a small amount of bias into the model estimates, ridge regression significantly reduces the variance, leading to a lower mean squared error (MSE) and better performance on new, unseen data.

---

## How It Works
The objective function minimized by ridge regression is:

    Minimize: Sum of Squared Residuals + λ(lambda) X (Sum of Squared Coefficients)

The λ (lambda) is the regularization parameter (or tuning parameter) that controls the strength of the penalty:

+ If λ = 0, the ridge model becomes standard linear regression (OLS).

+ As λ increases, the penalty becomes stronger, and the coefficients are shrunk more significantly towards zero.

Choosing an optimal λ value is crucial and is typically done using techniques like cross-validation to balance the bias-variance trade-off.

---

## When to Use a Ridge Model

Ridge regression is particularly useful in scenarios where:

1. `Multicollinearity is present`: It manages the high correlation between independent variables effectively.

2. `Overfitting needs to be addressed`: It reduces model complexity and helps the model generalize better to new data.

3. `All predictors are relevant`: Unlike Lasso regression (L1 regularization), which can eliminate features, ridge regression keeps all features in the model.

4. `The number of features is large relative to observations`: It can handle high-dimensional datasets well.

---

## Applications

+ `Healthcare`: Building diagnostic models with multiple medical factors.

+ `Engineering`: Predicting product failure probability from various sensor data inputs.

**K-Nearest Neighbors (KNN) regression is a non-parametric, instance-based machine learning algorithm used to predict continuous values. It operates on the principle that data points close to each other in the feature space likely have similar target values.
Instead of building a generalized model during a formal training phase, the KNN algorithm stores the entire dataset and performs computations only when a prediction for a new data point is needed (hence its "lazy learner" designation).**

---

## How KNN Regression Works

To predict the value for a new data point, the KNN algorithm follows these steps:

1. `Select k`: Choose the number of neighbors (k) to consider. The value of k is a crucial hyperparameter that affects model performance; a smaller k is sensitive to noise, while a larger k smooths predictions and reduces variance.

2. `Calculate Distances`: Calculate the distance (e.g., using Euclidean distance or Manhattan distance) between the new data point and all existing points in the training dataset.

3. `Find the K Nearest Neighbors`: Identify the k data points from the training set that are closest to the new data point.

4. `Predict the Value`: The final prediction for the new data point's target value is the average of the target values of its k nearest neighbors

---

## Advantages and Disadvantages

### Advantage
1. Simple and intuitive to understand and implement.

2. No training phase; new data can be added seamlessly.

3. Non-parametric; makes no assumptions about the data distribution, working well with non-linear data.

4. Versatile; can be used for both classification and regression tasks.

### Disadvantage

1. Computationally expensive during prediction because it calculates distances to all points.

2. High memory usage; requires storing the entire training dataset.

3. Prone to the curse of dimensionality; performance degrades as the number of features increases.

4. Sensitive to the scale of features; data needs to be properly scaled before use.

---

## Applications

KNN regression is used in various fields, including:

+ `Real Estate`: Predicting house prices based on the features and average prices of similar nearby houses.

+ `Finance`: Forecasting stock market trends or assessing loan risk by comparing applicant profiles to similar historical cases.

+ `Recommender Systems`: Suggesting products or movies to users based on the preferences of similar users.

+ `Medical Diagnosis`: Predicting a patient's likelihood of a certain condition by comparing their metrics with similar cases.

**Decision tree regression is a supervised machine learning algorithm that uses a tree-like structure to predict continuous numerical values, such as house prices or temperatures. It works by recursively splitting the dataset into smaller, more homogeneous subsets based on the input features until a stopping criterion is met.**

---

## How It Works

The algorithm begins at the root node with the entire dataset. It then iteratively finds the best split at each internal node by evaluating potential divisions across all features to maximize the reduction in prediction error. The data is partitioned based on this optimal split, and this process is repeated for the resulting child nodes. This recursive partitioning continues until stopping conditions are met, resulting in leaf nodes. For a new data point, the predicted value is the average of the target values of the training data points in the leaf node it reaches.

---

## Advantages and Disadvantages

### Advantage
1. Easy to understand and interpret due to its visual, flowchart-like structure.

2. Requires minimal data preparation.

3. Can handle both numerical and categorical data.

4. Can handle both numerical and categorical data.

### Disadvantage
1. Prone to overfitting, especially if the tree grows too deep, capturing noise in the data

2. High variance; small variations in the training data can result in a completely different tree structure.

3. Predictions are piecewise constant approximations.

4. Tends to be biased towards features with a large number of levels.

---

## Applications

Decision tree regression is applied in various areas, including financial forecasting, healthcare, marketing, and manufacturing. Techniques such as pruning or ensemble methods like Random Forest and Gradient Boosting are often used to improve performance and address the limitations of single decision trees.

**Support Vector Regression (SVR) is a robust machine learning algorithm used for predicting continuous values. It is an extension of the Support Vector Machine (SVM) algorithm, primarily known for classification, but adapted for regression tasks.
The core idea of SVR is to find the optimal hyperplane that fits the data points while maximizing the margin of tolerance, known as the epsilon (ε) margin. **

---

## Key Principles of SVR

1. `Hyperplane`: In SVR, the goal is to find a line (in 2D) or a hyperplane (in higher dimensions) that best approximates the relationship between input variables and the continuous target variable.

2. `ε-Insensitive Loss Function`: Unlike linear regression, which aims to minimize the error for every single data point, SVR uses an ε-insensitive loss function. Errors within a predefined "tube" (the ε-margin) around the hyperplane are simply ignored and not penalized.

3. `Support Vectors`: Only the data points that lie outside the ε-margin (the outliers or boundary points) influence the model's construction and the position of the hyperplane. These are called support vectors. This makes SVR particularly robust to outliers compared to traditional regression methods.

4. `Kernel Trick`: SVR can handle complex, non-linear relationships by using kernel functions (such as the Radial Basis Function (RBF), polynomial, or linear kernels). The kernel trick implicitly maps the data into a higher-dimensional feature space where a linear hyperplane can effectively solve the regression problem.

---

## Key Hyperparameters
SVR performance is tuned using specific hyperparameters:

1. `C (Regularization Parameter)`: Controls the trade-off between the flatness of the function (minimizing model complexity) and the tolerance for deviations larger than ε (penalizing errors). A smaller C allows for a wider margin and more errors, while a larger C enforces a stricter fit.

2. `ε (Epsilon)`: Defines the width of the tolerance tube around the regression line. It determines how much error is acceptable without penalty.

3. `Kernel`: Determines the type of hyperplane and the transformation of the data into a higher-dimensional space (e.g., 'linear', 'rbf', 'poly').

3. `Gamma (γ)`: A parameter for non-linear kernels like RBF, it controls the influence of individual training samples (support vectors); a low value means 'far' and a high value means 'close' influence.

---

## Applications
SVR is well-suited for high-dimensional data and cases with potential outliers. Common applications include:

+ Financial forecasting (e.g., stock price prediction)

+ Time series analysis

+ Healthcare (e.g., disease modeling, medical image analysis)

+ Engineering (e.g., predicting material properties or manufacturing outcomes)

**Random Forest Regression is a powerful ensemble machine learning algorithm that combines predictions from multiple decision trees to estimate a continuous numerical value. This method significantly improves accuracy and reduces overfitting compared to using a single decision tree by leveraging the "wisdom of the crowds".**

---

## How It Works

The magic of a random forest lies in the randomness injected during its creation, which helps decorrelate the individual trees.

+ `Bootstrap Aggregation (Bagging)`: The algorithm builds each decision tree using a random subset of the original training data sampled with replacement, meaning some data points may be used multiple times while others (known as out-of-bag samples) are left out.

+ `Feature Randomness`: At each node in a tree, only a random subset of features is considered when determining the best split point, ensuring diversity among the trees and preventing any single feature from dominating the model.

+ `Aggregation`: For a given input, each tree in the forest makes an independent prediction. The final output of the random forest regressor is the average (mean) of all these individual predictions.

---

## Advantages & Disadvantages

### Advantage

1. `High Accuracy`: Often produces very accurate predictions.

2. `Robustness`: Handles non-linear relationships, high-dimensional data, missing values, and outliers well.

3. `Prevents Overfitting`: The ensemble method and randomness significantly reduce the risk of overfitting.

4. `Feature Importance`: Provides valuable insights into which features are most important for predictions.

### Disadvantage

1. `Complexity`: Less interpretable (a "black box") compared to a single decision tree.

2. `Computational Cost`: Can be computationally expensive and slow to train and generate predictions, especially with many trees.

3. `Cannot Extrapolate`: It only predicts values within the range of the training data (averages of existing points), so it struggles with trends outside that range.

4. `Memory Intensive`: Storing multiple large trees requires significant memory resources.

---

## Common Applications

Random forest regression is used across various industries for continuous prediction tasks:

+ `Finance`: Predicting stock prices, assessing credit risk, and detecting fraud.

+ `Real Estate`: Estimating house prices based on features like size, location, and age.

+ `E-commerce`: Predicting customer lifetime value (CLV) and recommending products.

+ `Healthcare`: Analyzing patient data to predict drug responses or identify disease risk factors.
